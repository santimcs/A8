```python
import pandas as pd
from datetime import date, timedelta
from sqlalchemy import create_engine

engine = create_engine('sqlite:///c:\\ruby\\portlt\\db\\development.sqlite3')
conlt = engine.connect()

engine = create_engine('postgresql+psycopg2://postgres:admin@localhost:5432/portpg_development')
conpg = engine.connect()

today = date.today()
today
```

```python
sql = '''
SELECT name, id AS ticker_id
FROM tickers'''

tickers = pd.read_sql(sql, conpg)
tickers.shape
```

```python
sql = '''
DELETE FROM epss'''
rp = conpg.execute(sql)
rp.rowcount
```

```python
sql = '''
SELECT name,year,quarter,q_amt,y_amt,aq_amt,ay_amt,q_eps,y_eps,aq_eps,ay_eps,publish_date
FROM epss 
ORDER BY year, quarter, name'''

df_inp = pd.read_sql(sql, conlt)
df_inp.shape
```

```python
df_merge = pd.merge(df_inp, tickers, on='name', how='outer', indicator = True)
df_merge.shape
```

```python
df_left = df_merge[df_merge['_merge'] == 'left_only']
df_left['name'].unique()
df_left
```

```python
df_ins = df_merge[df_merge['_merge'] == 'both']
df_eps = df_ins[['name','year','quarter',\
     'q_amt','y_amt',\
     'aq_amt','ay_amt',\
     'q_eps','y_eps',\
     'aq_eps','ay_eps','ticker_id','publish_date']]
df_eps.shape
```

```python
rcds = df_eps.values.tolist()
len(rcds)
```

```python
for rcd in rcds:
    print(rcd)
```

```python
for rcd in rcds:
    conpg.execute("""INSERT INTO epss (name, year, quarter, \
q_amt, y_amt, aq_amt, ay_amt, \
q_eps, y_eps, aq_eps, ay_eps, ticker_id, publish_date) \
VALUES( %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)""", rcd) 
```

### Start of Yearly Profit Section

```python
sql = '''
DELETE FROM yr_profits'''
rp = conpg.execute(sql)
rp.rowcount
```

```python
sql = '''
SELECT name, year, quarter, latest_amt, previous_amt, inc_amt, inc_pct 
FROM yr_profits 
ORDER BY year desc, quarter desc, name'''
df_year = pd.read_sql(sql, conlt)
df_year.shape
```

```python
df_year['qtr_int'] = df_year['quarter'].str[1:]
df_year.shape
```

```python
df_merge = pd.merge(df_year, tickers, on='name', how='outer', indicator = True)
df_merge.shape
```

```python
df_left = df_merge[df_merge['_merge'] == 'left_only']
df_left
```

```python
df_ins = df_merge[df_merge['_merge'] == 'both']

df_eps = df_ins[['name','year','qtr_int',\
    'latest_amt','previous_amt',\
    'inc_amt','inc_pct','ticker_id']]
df_eps.shape
```

```python
rcds = df_eps.values.tolist()
len(rcds)
```

```python
for rcd in rcds:
    print(rcd)
```

```python
for rcd in rcds:
    conpg.execute("""INSERT INTO yr_profits (name, year, quarter, \
latest_amt, previous_amt, inc_amt, inc_pct, ticker_id) \
VALUES( %s, %s, %s, %s, %s, %s, %s, %s)""", rcd)
```

### Start of Profits section

```python
sql = '''
SELECT name, id AS ticker_id
FROM tickers'''
tickers = pd.read_sql(sql, conpg)
tickers.shape
```

```python
sql = '''
DELETE FROM profits'''
rp = conpg.execute(sql)
rp.rowcount
```

```python
sql = '''
SELECT *
FROM profits 
ORDER BY year desc, quarter desc, name'''
profits = pd.read_sql(sql, conlt)
profits.shape
```

```python
sql = '''
SELECT name, year, quarter, publish_date
FROM epss'''
epss = pd.read_sql(sql, conlt)
epss.shape
```

```python
df_merge = pd.merge(profits, epss, on=['name','year','quarter'], how='outer', indicator = True)
df_merge.shape
```

```python
prf_eps = df_merge[df_merge['_merge'] == 'both']
prf_eps.shape
```

```python
columns = ['id','ticker_id','_merge']
prf_eps_2 = prf_eps.drop(columns, axis=1)
prf_eps_2.shape
```

```python
df_merge = pd.merge(prf_eps_2, tickers, on='name', how='inner')
df_merge.shape
```

```python
rcds = df_merge.values.tolist()
len(rcds)
```

```python
for rcd in rcds:
    print(rcd)
```

```python
for rcd in rcds:
    conpg.execute("""INSERT INTO profits (name, year, quarter, kind,\
latest_amt_y, previous_amt_y, inc_amt_y, inc_pct_y,\
latest_amt_q, previous_amt_q, inc_amt_q, inc_pct_q,\
q_amt_c, y_amt, inc_amt_py, inc_pct_py,\
q_amt_p, inc_amt_pq, inc_pct_pq,\
mean_pct, std_pct, publish_date, ticker_id)\
VALUES( %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)""", rcd)
```
