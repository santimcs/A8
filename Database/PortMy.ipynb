```python
sql = """
SELECT P.name, P.ticker_id, T.name, T.id
FROM portfolios P
JOIN tickers T ON P.name = T.name
ORDER BY P.name
"""
portfolios = pd.read_sql(sql, conmy)
portfolios
```
