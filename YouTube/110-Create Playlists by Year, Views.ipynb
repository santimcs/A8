```python
import pandas as pd
import datetime as dt
from sqlalchemy import create_engine

engine = create_engine('postgresql+psycopg2://postgres:admin@localhost:5432/BBB_development')
conbb = engine.connect()
    
data_path = '../data/'
```

```python
fm_year = 1970
to_year = 1970
sql = '''
SELECT year, rank, nbr, title, A.name, code, publish_date, views
FROM songs S
JOIN artists A ON artist_id = A.id
WHERE    code <> '' AND year BETWEEN %s AND %s
ORDER BY views DESC LIMIT 12'''
sql = sql % (fm_year, to_year)
songs = pd.read_sql(sql,conbb, parse_dates = ["publish_date"])
songs
```

```python
sql = '''
SELECT SUM(views)
FROM songs S
JOIN artists A ON artist_id = A.id
WHERE year BETWEEN %s AND %s
AND   code <> '' LIMIT 12
'''
sql = sql % (fm_year, to_year) 
pd.read_sql(sql,conbb)
```
