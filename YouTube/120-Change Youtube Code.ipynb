```python
from bs4 import BeautifulSoup
import requests
```

```python
inp_file = 'b7906.html'
url = 'https://www.santiwebsite.com/songs/' + inp_file
url
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup)
```
