```python
import requests
from bs4 import BeautifulSoup
```

```python
# Get the website
url = 'https://subslikescript.com/movies'
page = requests.get(url)
page
```

```python
# Parse the HTML
soup = BeautifulSoup(page.content, 'html.parser')
soup
```

```python
# Find the element
ul_element = soup.find('ul', {'class': 'scripts-list'})
ul_element
```

```python
# Get all 'a' elements
a_elements = ul_element.find_all('a')
a_elements
```

```python
# Print the text attribute of each element
for element in a_elements:
    print(element.text)
```
