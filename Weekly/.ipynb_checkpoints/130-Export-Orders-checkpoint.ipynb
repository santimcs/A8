### After call Ord-Log.bat

```python
file_name   = 'orders-log.csv'
input_file = data_path + file_name
col_names   = ['trade','name','spreads','reason','market','qty','price','cur_price','chg_amt','chg_pct','diff','active']
df_log = pd.read_csv(input_file, sep=',', header=None, index_col=None, names=col_names)
df_log
```

```python
def calc(vals):
    spreads = vals
    if abs(spreads) < 10:
        return 1
    else:
        return 0
```

```python
df_log['active'] = df_log['spreads'].apply(calc)
```

```python
df_log.groupby('active')['trade'].count()
```

```python
sql = '''
DELETE FROM orders'''
rp = conlite.execute(sql)
rp.rowcount
```

```python
final = df_log[['trade','name','qty','price','active','reason','market']]
rcds = final.values.tolist()
len(rcds)
```

```python
for rcd in rcds:
    print(rcd)
```

```python
sql = '''
SELECT *
FROM orders
ORDER BY trade, name'''
df_orders = pd.read_sql(sql, conlite)
df_orders.columns
```

```python
df_tab = pn.widgets.Tabulator(df_orders, layout='fit_data', width=800)
df_tab
```
