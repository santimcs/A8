```python
import pandas as pd
from datetime import date, timedelta
from sqlalchemy import create_engine

engine = create_engine("sqlite:///c:\\ruby\\portmy\\db\\development.sqlite3")
conmy = engine.connect()
engine = create_engine("postgresql+psycopg2://postgres:admin@localhost:5432/portpg_development")
conpg = engine.connect()
engine = create_engine("mysql+pymysql://root:@localhost:3306/stock")
const = engine.connect()

data_path = "../data/"
csv_path = "\\Users\\User\\iCloudDrive\\"
box_path = "\\Users\\User\\Dropbox\\"

pd.set_option('display.max_row', None)
format_dict = {'volume':'{:,}',
               'dividend':'{:.2f}','price':'{:.2f}','target':'{:.2f}','unit_cost':'{:.2f}',
               'price_d':'{:.2f}','price_w':'{:.2f}',
               'max':'{:.2f}','min':'{:.2f}','pe':'{:.2f}','pbv':'{:.2f}','beta':'{:.2f}',               
               'cost_amt':'{:,.2f}','div_amt':'{:,.2f}','yield':'{:,.2f}','yld':'{:,.2f}',
               'pct':'{:,.2f}%'}

today = date.today()
week_str = today.strftime('%Y-%U')
today, week_str
```

```python
today = today - timedelta(days=1)
wkago = today - timedelta(days=8)
today, wkago
```

```python
sql = '''
SELECT name, kind, year, quarter
FROM profits
ORDER BY name'''
my_profits = pd.read_sql(sql, conmy)
my_profits.shape[0]
```

```python
sql = """
SELECT name, price, target_price AS target, 
ROUND((target_price-price)/price*100, 0) AS upside, buy, hold, sell, yield, 
date(updated_at), ('%s'::date - date(updated_at)::date) AS days
FROM consensus
WHERE price > 0 AND target_price > 0
AND ('%s'::date - date(updated_at)::date) <= 31"""
sql = sql % (today, today)
consensus = pd.read_sql(sql, conpg)
consensus.set_index('name', inplace=True)
consensus['upside'] = consensus['upside'].astype(int)
consensus.shape[0]
```

```python
consensus.loc['TSTH']
```

```python
consensus.loc['TSTH',['target','upside']] = [1.89,21]
```

```python
prf_css = pd.merge(my_profits, consensus, on='name', how='inner')
prf_css.set_index('name', inplace=True)
prf_css.shape[0]
```

### Profits w/o consensus

```python
df_tmp = pd.merge(my_profits, consensus, on='name', how='outer',indicator=True)
prf_wo_css = df_tmp['_merge'] == 'left_only'
df_tmp[prf_wo_css]
```

### Start of Gain Percentage Calculation

```python
sql = '''
SELECT name, max_price AS max, min_price AS min, pe, pbv, daily_volume AS volume, beta, market
FROM stocks
'''
my_stocks = pd.read_sql(sql, conmy)
my_stocks.shape[0]
```

```python
consensus2 = pd.merge(prf_css, my_stocks, on='name', how='inner')
consensus2.set_index('name', inplace=True)
consensus2.shape
```

```python
set50 = consensus2.market.str.contains('SET50') & (consensus2.upside >= 5)
flt_set50 = consensus2[set50]
flt_set50.sort_values(['name'],ascending=[True]).shape[0]
```

```python
consensus2.loc[
    consensus2.market.str.contains('SET50') & (consensus2.upside >= 5),
    ['year', 'quarter', 'price', 'target', 'upside', 'buy', 'hold', 'sell', 'yield', 'max', 'min', 'pe', 'pbv', 'volume', 'beta']
]
```

```python
set100 = consensus2.market.str.contains('SET100') & (consensus2.upside >= 10)
flt_set100 = consensus2[set100]
flt_set100.shape[0]
```

```python
consensus2.loc[
    consensus2.market.str.contains('SET100') & (consensus2.upside >= 10),
    ['year', 'quarter', 'price', 'target', 'upside', 'buy', 'hold', 'sell', 'yield', 'max', 'min', 'pe', 'pbv', 'volume', 'beta']
]
```

```python
setmai = consensus2.market.str.contains('mai') & (consensus2.upside >= 15)
flt_setmai = consensus2[setmai]
flt_setmai.sort_values(['name'],ascending=[True]).style.format(format_dict)
```

```python
flt_set999 = consensus2[~(set100 | set50 | setmai) & 
            (consensus2.upside >= 15) ]
flt_set999.shape[0]
```

```python
consensus2.loc[
    ~(set100 | set50 | setmai) & (consensus2.upside >= 15),
    ['year', 'quarter', 'price', 'target', 'upside', 'buy', 'hold', 'sell', 'yield', 'max', 'min', 'pe', 'pbv', 'volume', 'beta']
]
```

```python
flt_all = pd.concat([flt_set50,flt_set100,flt_set999,flt_setmai])
flt_all.sort_values(['upside'],ascending=[False])
```

```python
flt_all.shape
```

```python
sql = '''
SELECT name, year, quarter
FROM charts 
ORDER BY name, year DESC, quarter DESC
'''
charts = pd.read_sql(sql, conpg)
charts.shape
```

### Start of Delete Duplicated Charts

```python
df_uniq = charts.drop_duplicates(subset='name', keep='first')
df_uniq.shape[0]
```

```python
sr = df_uniq['name']
sr
```

```python
names = sr.values.tolist()
in_p = ', '.join(map(lambda name: "'%s'" % name, names))
in_p
```

### End of Delete Duplicated Charts

```python
filter = pd.merge(df_uniq, flt_all, on='name', how='inner')
filter.sort_values(by=['name'],ascending=[True]).style.format(format_dict)
```

```python
final = filter
col9 = 'name price target upside buy hold sell market yield'.split()
final[col9].sort_values(by='name',ascending=True).style.format(format_dict)
```

```python
sql = '''
SELECT name, sector, subsector
FROM tickers'''
pg_tickers = pd.read_sql(sql, conpg)
pg_tickers.shape[0]
```

```python
final2 = pd.merge(final, pg_tickers, on='name', how='inner')
colt = 'name price target upside buy hold sell market sector subsector volume beta yield'.split()
final2.reset_index()
final2[colt].sort_values(['name'],ascending=[True]).to_json("C:/ClearStep/dist/consensus.json", orient="table")
```
