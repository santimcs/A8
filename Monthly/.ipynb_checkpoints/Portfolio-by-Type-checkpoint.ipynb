```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
from sqlalchemy import create_engine
from datetime import date, timedelta
import calendar

pd.options.display.max_rows = 8

engine = create_engine('mysql+pymysql://root:@localhost:3306/stock')
conmy = engine.connect()
```

```python
month = date(2019, 5, 1)
bom = month
eom = month.replace(day=calendar.monthrange(month.year, month.month)[1])
bom, eom
```

```python
eom = date(2019, 5, 31)
```

```python
sql = '''
SELECT B.name, B.date, volbuy, volbuy * B.price AS cost_amount, 
volbuy * P.price AS market_value, period 
FROM buy B JOIN price P USING (name) 
WHERE active=1 AND P.date = "%s" 
ORDER BY period, B.name'''
sql = sql % eom
print(sql)
```

```python
df = pd.read_sql(sql, conmy, parse_dates=['date'], index_col='name')
df.shape
```

```python
file_name   = 'stock-by-grade.csv'
output_file = '../data/' + file_name
df_out = df
df_out.to_csv(output_file)
```

```python
df.period.replace(['1','2','3','4'], ['disposal','high dividend','long term','short term'], inplace=True)
grouped = df.groupby(by='period')
grouped['cost_amount','market_value'].sum()
```

```python
cost_summary = grouped.sum().cost_amount
cost_summary
```

```python
sum(cost_summary)
```

```python
fig_path = '../fig/Portfolio-at-Cost-1905.png'
with plt.style.context('my_custom_style'):
    plt.figure(figsize=(12, 12))
    ax = cost_summary.plot.pie(colors=['green','red','orange','cyan'],
                               startangle=90,
                               autopct='%1.1f%%')
    plt.savefig(fig_path)
```

```python
market_summary = grouped.sum().market_value
market_summary
```

```python
sum(market_summary)
```

```python
fig_path = '../fig/Portfolio-at-Market-1905.png'
with plt.style.context('my_custom_style'):
    plt.figure(figsize=(12, 12))
    ax = market_summary.plot.pie(colors=['green','red','orange','cyan'],
                               startangle=90,
                               autopct='%1.1f%%')
    plt.savefig(fig_path)
```
