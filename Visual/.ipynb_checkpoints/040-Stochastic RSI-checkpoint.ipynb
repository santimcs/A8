```python
import pandas as pd
import numpy as np
from datetime import date, timedelta
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
# do all our graphics inline
%matplotlib inline
# Set some pandas options
pd.set_option('display.notebook_repr_html', False)
pd.set_option('display.max_columns', 8)
pd.set_option('display.max_rows', 10) 
pd.set_option('display.width', 78) 
pd.set_option('precision', 3)

engine = create_engine("mysql+pymysql://root:@localhost:3306/stock")
const = engine.connect()

today = date.today()
one_year_ago = today - timedelta(days=180)
today, one_year_ago
```

### Get past one year data

```python
sql = """
SELECT date AS Date, maxp AS High, minp AS Low, opnp AS Open, price AS Close,
       qty AS Volume, price AS 'Adj Close'
FROM price 
WHERE date >= '%s' AND name = '%s' 
ORDER BY Date"""
sql = sql % (one_year_ago, name)
print(sql)
```

```python
df = pd.read_sql(sql, const)
df = df.set_index(pd.DatetimeIndex(df['Date'].values))
df[:5]
```

```python
plt.figure(figsize=(16,8))
plt.title(name)
plt.plot(df['Close'])
plt.xlabel('Date')
plt.ylabel('Close Price in Baht')
plt.show()
```

```python
def EMA(data, period=20, column='Close'):
    return data[column].ewm(span=period, adjust=False).mean()
```

```python
def StochRSI(data, period=14, column='Close'):
    delta = data[column].diff(1)
    #delta = delta.dropna()    
    up = delta.copy()
    down = delta.copy()   
    up[up<0] = 0
    down[down>0] = 0     
    data['up'] = up
    data['down'] = down
    AVG_Gain = EMA(data, period, column='up')
    AVG_Loss = abs( EMA(data, period, column='down') )
    RS = AVG_Gain / AVG_Loss
    RSI = 100.0 - (100.0/(1.0 + RS))
    
    stockrsi = (RSI - RSI.rolling(period).min()) / (RSI.rolling(period).max() - RSI.rolling(period).min())

    return stockrsi
```

```python
df['StochRSI'] = StochRSI(df)
```

```python
#plt.figure(figsize=(18,9))
fig, (ax1, ax2) = plt.subplots(nrows=2, sharex=True)
plt.subplots_adjust(hspace=.1)
plt.title(name, fontsize=18)
#ax1.grid()
#ax2.grid()
ax1.plot(df.index, df['Close'], color='r')
ax2.plot(df.index, df['StochRSI'], color='b', linestyle='--')
ax2.axhline(0.20, color='orange')
ax2.axhline(0.80, color='orange')
plt.xticks(rotation=45)
plt.gcf().set_size_inches(12,9)
```
