```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.bbc.com/sport/football/tables'
r = requests.get(url)
html = r.text
soup = BeautifulSoup(html, 'html.parser')
table = soup.find('table')
df = pd.read_html(str(table))[0]
df
```

```python
df.drop(columns=['Unnamed: 0','Unnamed: 1','Form'],inplace=True)
df.drop([20],inplace=True)
df
```
