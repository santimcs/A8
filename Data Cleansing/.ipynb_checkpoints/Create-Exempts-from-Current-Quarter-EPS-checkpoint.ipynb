```python
import pandas as pd
from sqlalchemy import create_engine
from glob import glob

engine = create_engine("sqlite:///c:\\ruby\\portlt\\db\\development.sqlite3")
conlt = engine.connect()

engine = create_engine('mysql+pymysql://root:@localhost:3306/portfolio_development')
conpf = engine.connect()

engine = create_engine(
    "postgresql+psycopg2://postgres:admin@localhost:5432/portpg_development"
)
conpg = engine.connect()

pd.set_option('max_rows',None)

data_path = "../data/"
```

```python
year = 2021
quarter = 3
year, quarter
```

```python
sql = """
SELECT name, year, quarter, q_amt, y_amt, aq_amt, ay_amt, ticker_id
FROM epss
WHERE year = %s AND quarter = %s
ORDER BY name
"""
sql = sql % (year, quarter)
epss = pd.read_sql(sql, conlt)
epss.shape
```

```python
epss['grade'] = 0
epss.columns
```

```python
def grade(vals):
    qc, qp, aqc, aqp = vals
    if aqc >= aqp:
        if qc >= qp:
            return 1
        else:
            return 2
    else:
        if qc < qp:
            return 4
        else:
            return 3
```

```python
epss["grade"] = epss[["q_amt", "y_amt","aq_amt", "ay_amt"]].apply(grade, axis=1)
epss.shape
```

```python
mask1 = epss["grade"] == 1
epss[mask1].shape
```

```python
mask2 = epss["grade"] == 2
epss[mask2].shape
```

```python
mask3 = epss["grade"] == 3
epss[mask3].shape
```

```python
mask4 = epss["grade"] == 4
epss[mask4].shape
```

```python
g4 = epss[mask4]
g4.sort_values(by='name',ascending=True)
```

```python
mask41 = g4.aq_amt < 200000
g4[mask41].shape
```

```python
mask42 = g4.q_amt < 100000
g4[mask42].shape
```

```python
temp = g4[mask41 & ~mask42]
temp.sort_values(by='name',ascending=True)
```

```python
filter1 = g4[mask41 & mask42]
filter1.sort_values(by='name',ascending=True).shape
```
