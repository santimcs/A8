```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
from sqlalchemy import create_engine

engine1 = create_engine("mysql+pymysql://root:@localhost:3306/stock")
const = engine.connect()
```

```python
df = pd.read_csv("../data/div-20q4.csv", parse_dates=['xd_date','pay_date'],
                dtype={'name':'string','shares':'int'},nrows=41)
df.dtypes
```

```python
df
```

```python
df.xd_date = df.xd_date + timedelta(days=365)
df.pay_date = df.pay_date + timedelta(days=365)
df.xd_date,df.pay_date
```

```python
df.to_sql('div_update', engine1, if_exists='replace')
```

```python
sql = '''
SELECT *
FROM div_update'''
df_upd = pd.read_sql(sql, const)
df_upd
```

```python
sqlUpd = '''
UPDATE dividend D
SET XDATE = 
(SELECT xd_date FROM div_update U
WHERE D.NAME = U.name)'''
print(sqlUpd)
```

```python
rp = const.execute(sqlUpd)
rp.rowcount
```

```python
sql = '''
SELECT * FROM dividend WHERE XDATE IS NULL
'''
df_dividend = pd.read_sql(sql, const)
df_dividend
```

```python
sqlUpd = '''
UPDATE dividend
SET XDATE = PAIDDATE WHERE XDATE IS NULL
'''
rp = const.execute(sqlUpd)
rp.rowcount
```
